```python
import pandas as pd


# 创建一个空的DataFrame来存储结果
artist_df = pd.DataFrame(columns=['en_name','cn_name','link'])

for index, row in artists.iterrows():
    link = row['Link']
    name = row['Name']
    
    # Check if the result DataFrame already has this link
    if link in artist_df['link'].values:
        # If link exists, find the index and update 'Name2'
        idx = artist_df[artist_df['link'] == link].index[0]
        artist_df.at[idx, 'cn_name'] = name
    else:
        # If link does not exist, create a new DataFrame row and concatenate it
        new_row_df = pd.DataFrame([[name, pd.NA, link]], columns=['en_name', 'cn_name', 'link'])
        artist_df = pd.concat([artist_df, new_row_df], ignore_index=True)

artist_df.index = artist_df.index + 1        
artist_df.to_csv(city_name+'_artists.csv',index=True,sep=',')       
artist_df.head(50)
```
